#Gráfico de linea

```python
import matplotlib.pyplot as plt
 
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.plot(x, y1, label="linea I")
plt.plot(x, y2, label="linea H")
plt.plot()

plt.xlabel("x")
plt.ylabel("y")
plt.title("Grafico de lineas")
plt.legend()
plt.show()
```

#Gráfico de barras #2

```python
import matplotlib.pyplot as plt

x1 = [1, 3, 4, 5, 6, 7, 9]
y1 = [4, 7, 2, 4, 7, 8, 3]

x2 = [2, 4, 6, 8, 10]
y2 = [5, 6, 2, 6, 2]

plt.show()

plt.bar(x1, y1, label="Verde", color='g')
plt.bar(x2, y2, label="Amarillo", color='aquamarine')
plt.plot()

plt.xlabel("Numero")
plt.ylabel("Altura")
plt.title("Grafico de barras")
plt.legend()
plt.show()
```

#Gráficos de Histogramas

```python
import matplotlib.pyplot as plt
import numpy as np

n = np.random.randn(1000)

m = [m for m in range(len(n))]
plt.bar(m, n, color='g')
plt.title("Datos de calificaciones")
plt.show()

plt.hist(n, bins=20, color='g')
plt.title("Histograma")
plt.show()
```

# Gráficos con marcas

```python
import matplotlib.pyplot as plt

x1 = [2, 3, 4]
y1 = [5, 5, 5]

x2 = [1, 2, 3, 4, 5]
y2 = [2, 3, 2, 3, 4]
y3 = [6, 8, 7, 8, 7]

plt.scatter(x1, y1, marker="^", color='c')
plt.scatter(x2, y2, marker="<", color='y')
plt.scatter(x2, y3, marker='>', color='m')
plt.title('Gráfico de dispersión')
plt.show()
```
